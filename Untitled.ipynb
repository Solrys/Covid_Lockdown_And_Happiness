```python
import pandas as pd
```

```python
happiness_df = pd.read_csv("2020.csv")
happiness_df.head()
```

```python
new_lock_df = pd.read_csv("new_lock_df.csv")
new_lock_df.head()
```

```python
# Rename Country Column to match
lock_df = new_lock_df.rename(columns={'Country/Region': 'Country name'})
lock_df.head()
```

```python
# Combine the data into a single dataset.
happy_lock_df = pd.merge(lock_df, happiness_df, on=["Country name", "Country name"])
happy_lock_df.head()
```

```python
#Save New File
output_file_path = "happy_lock_df.csv"
happy_lock_df.to_csv(output_file_path, index=False)
```
